```python
url= ["http://leagueoflegends.wikia.com/wiki/Category:Assassin_champion",
      "http://leagueoflegends.wikia.com/wiki/Category:Carry_champion",
      "http://leagueoflegends.wikia.com/wiki/Category:Fighter_champion",
      ""]
```

```python


dic = {}
for u in url:
    CCType = re.search(r"y:(.*)", u).group(1)
    page=urllib2.urlopen(u)
    soup = BeautifulSoup(page.read())
    champs = []
    for li in soup('li'):
        for a in li.findAll('a',text = re.compile('/Ability'),attrs={'class': None}):
            name = a.replace('/Ability Details','')
            if name not in champs:
                champs.append(name)
    dic[CCType] = champs
```
