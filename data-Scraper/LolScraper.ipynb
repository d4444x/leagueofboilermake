```python
def getMatchInfoDict(match):
    result = {}
    for d in match:
        result[d['name']] = d['value']
    return result

def parseMatches(matches, summonerId):
    
    gamesToSave = []
    
    # setup the list of playerIDs that we have to spider
    summonerIds = []
    
    for match in matches:
        
        # Setup the dict for the 2 teams
        championsInMatch = {}
        
        # Setup the list for the champs for each team
        championsInMatch[100] = []
        championsInMatch[200] = []
        
        # Add the current player to his respective team
        championsInMatch[match['teamId']].append(match["championId"])
        
        if not ('fellowPlayers' in match):
            # pretty sure that we dont want this data...
            continue
        
        # Add each other player in the match to their team
        for player in match['fellowPlayers']:
            championsInMatch[player["teamId"]].append(player["championId"])
            
            # Store the list of summoners that we have to lookup for this match
            summonerIds.append(player["summonerId"])
                    
        matchInfo = getMatchInfoDict(match['statistics'])
        
        # If the key WIN is present, that means that the current player won
        if ('WIN' in matchInfo):
            winner = matchInfo['TEAM']
        # The key LOSE is present, current player lost
        elif ('LOSE' in matchInfo):
            winner = (200 if matchInfo['TEAM']==100 else 100)
        else:
            winner = 0
            
        championsInMatch[100].sort()
        championsInMatch[200].sort()
        
        gameToSave = {"gameId":match["gameId"],
                      "date":datetime.datetime.utcfromtimestamp(int(str(match['createDate'])[:-3])),
                      "champions_100":championsInMatch[100],
                      "champions_200":championsInMatch[200],
                      "winner":winner,
                      "mapId":match["mapId"]}
        
        gamesToSave.append(gameToSave)
        
    return summonerIds,gamesToSave
       
def get_games(id):
    errorCount = 0
    while(True):
        try:
            return lolApi.get_game_by_id(id)
        except urllib2.HTTPError as e:
            if e.code == 429:
                print "429'ed, waiting..."
                time.sleep(5)
                
            else:
                errorCount += 1
                if errorCount > 3:
                    print "Retry Failed..."
                    raise
                print 'Retrying, Attempt: ' + str(errorCount)
                time.sleep(10)
            
def get_games_of_seed(name):
    errorCount = 0
    while(True):
        try:
            return lolApi.get_game(name)
        except urllib2.HTTPError as e:
            if e.code == 429:
                print "429'ed, waiting..."
                time.sleep(5)
            else:
                errorCount += 1
                if errorCount > 3:
                    print "Retry Failed..."
                    raise
                print 'Retrying, Attempt: ' + str(errorCount)
                time.sleep(10)
```
